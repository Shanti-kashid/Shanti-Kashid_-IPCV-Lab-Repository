```python
img = cv2.imread(user_provided_path, 0)   # 0 = grayscale
print(img.shape)
```

```python
if 'img' not in locals() or img is None:
    # Assuming user_provided_path is defined in an earlier cell
    # If user_provided_path is also not defined, this will raise a NameError
    img = cv2.imread(user_provided_path, 0)

if img is not None:
    plt.imshow(img.astype(np.uint8), cmap='gray')
    plt.title("Original Image")
    plt.axis("off")
    plt.show()
else:
    print(f"Error: Image could not be loaded or is None for path: {user_provided_path}")
```

```python
hist = np.zeros(256)
hist
```

```python
for pixel in img.flatten():
    hist[pixel] += 1

print(hist)
```

```python
total_sum = img.sum()
print(total_sum)
```

```python
hist = hist / img.size
print(hist)
```

```python
cdf = np.cumsum(hist)
print(cdf)

#np.cumsum: Short for "Cumulative Sum."
#cdf = total number of pixels
```

```python
new_values = np.round(255 * cdf).astype(np.uint8)
print(new_values)
```

```python
equalized_img = new_values[img]
```

```python
plt.imshow(equalized_img, cmap='gray')
plt.title("Equalized Image")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original")

plt.subplot(1,2,2)
plt.imshow(equalized_img, cmap='gray')
plt.title("Equalized")

plt.show()
```

```python
eq2 = cv2.equalizeHist(img)

plt.imshow(eq2, cmap='gray')
plt.title("OpenCV Equalized")
plt.axis("off")
plt.show()
```

Let's also visualize the PDF for the equalized image to see the effect of histogram equalization.

```python
import numpy as np
import matplotlib.pyplot as plt

# Calculate histogram for the equalized image
equalized_hist, bins = np.histogram(equalized_img.flatten(), 256, [0,256])
equalized_hist = equalized_hist / equalized_img.size

plt.figure(figsize=(10, 5))
plt.plot(equalized_hist)
plt.title('Probability Density Function (Normalized Histogram) of Equalized Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```
